```python
import mysql.connector
import pandas as pd
import sqlalchemy
import numpy as np
conn=sqlalchemy.create_engine(
    'mysql+pymysql://root:makanayam@localhost:3306/world'
)
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
query='''
select * from country
where Region='Southeast Asia'
order by Name
'''

df=pd.read_sql_query(sqlalchemy.text(query),conn)
df
```

```python
sns.set(style='darkgrid')
sns.barplot(x=df.Name,y=df.Population,palette='muted')
plt.xlabel('Negara')
plt.ylabel('Populasi (x 1000Jt Jiwa)')
plt.title('Populasi Negara ASEAN')
plt.xticks(rotation=65, fontsize=9)
xlocs = plt.xticks()
for i, y in enumerate(list(df.Population)):
    plt.text(xlocs[0][i] - 0.30, y + 1000000, str(y), fontsize=6)
plt.grid(True)
plt.show()
```

```python
plt.style.use('seaborn')
color = sns.color_palette('muted')
_, _, autotexts = plt.pie(df.SurfaceArea,labels=df.Name,autopct='%1.1f%%',colors=color)
for autotext in autotexts:
    autotext.set_color('white')
plt.title('Persentase Penduduk ASEAN')
plt.axis('equal')
plt.show()
```

```python
sns.set(style='darkgrid')
sns.barplot(x=df.Name,y=df.GNP,palette='muted')
plt.xlabel('Negara')
plt.ylabel('Gross National Product (US$)')
plt.title('Pendapatan Bruto Nasional ASEAN')
plt.xticks(rotation=65, fontsize=9)
xlocs = plt.xticks()
for i, y in enumerate(list(df.GNP)):
    plt.text(xlocs[0][i] - 0.30, y, str(y), fontsize=7)
plt.grid(True)
plt.show()
```

```python
plt.style.use('seaborn')
color = sns.color_palette('muted')
_, _, autotexts = plt.pie(df.SurfaceArea,labels=df.Name,autopct='%1.1f%%',colors=color)
for autotext in autotexts:
    autotext.set_color('white')
plt.title('Persentase Luas Daratan ASEAN')
plt.axis('equal')
plt.show()
```
